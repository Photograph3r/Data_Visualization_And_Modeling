```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = px.data.gapminder()
df.head()
```

```python
df_sorted = df.sort_values('year')
fig = px.choropleth(df_sorted,
    locations='iso_alpha',
    color='gdpPercap',
    hover_name='country',
    animation_frame='year',
    title='GDP Per Capita by Country Over Time',
    color_continuous_scale='Viridis')
fig.show()
```

```python
df_sorted2 = df.sort_values('gdpPercap')
fig2 = px.choropleth(df_sorted2,
    locations='iso_alpha',
    color='lifeExp',
    hover_name='country',
    animation_frame='gdpPercap',
    title='Life Expectancy Animated by GDP Per Capita',
    color_continuous_scale='Viridis')
fig2.show()
```

```python
df_sorted3 = df.sort_values('year')
fig3 = px.choropleth(df_sorted3,
    locations='iso_alpha',
    color='pop',
    hover_name='country',
    animation_frame='year',
    title='Population by Country Over Time',
    color_continuous_scale='Viridis')
fig3.show()
```

## Which Visualization Best Illustrates Trends?

The visualization that best illustrates trends in GDP, life expectancy, and 
population is the **GDP per capita choropleth animated by year** (Cell 2).

It clearly shows how wealth has grown over time across different regions, 
with Western Europe and North America consistently leading while Sub-Saharan 
Africa has lagged behind. Animating by year gives a natural sense of progression 
that is easy to follow and interpret.

The life expectancy animation by GDP is harder to interpret because GDP is not 
evenly spaced, making the animation feel uneven and harder to follow.
