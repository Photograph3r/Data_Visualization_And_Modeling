## Visualization 1: Average Salary by Country Over Time

```python
salary_map = df.groupby(['country', 'year'])['salary_usd_annual'].mean().reset_index()
salary_map['iso_alpha'] = salary_map['country'].map(iso_map)
salary_map = salary_map.dropna(subset=['iso_alpha'])
salary_map = salary_map.sort_values('year')

fig1 = px.choropleth(salary_map,
    locations='iso_alpha',
    color='salary_usd_annual',
    hover_name='country',
    animation_frame='year',
    title='Average Salary (USD) by Country Over Time (2024-2026)',
    color_continuous_scale='Blues')
fig1.show()
```

This choropleth map shows how average annual salaries differ across countries 
from 2024 to 2026. North America, Western Europe, and Australia consistently 
show the highest salaries. Emerging markets like India, Pakistan, and parts of 
Africa show lower average salaries. The animation reveals subtle shifts in 
salary levels year over year globally.

## Visualization 2: Female Salary by Country Over Time

```python
gender_salary = df.groupby(['country', 'year', 'gender'])['salary_usd_annual'].mean().reset_index()
gender_salary['iso_alpha'] = gender_salary['country'].map(iso_map)
gender_salary = gender_salary.dropna(subset=['iso_alpha'])

female_salary = gender_salary[gender_salary['gender'] == 'Female'].sort_values('year')

fig2 = px.choropleth(female_salary,
    locations='iso_alpha',
    color='salary_usd_annual',
    hover_name='country',
    animation_frame='year',
    locationmode='ISO-3',
    title='Average Female Salary (USD) by Country Over Time (2024-2026)',
    color_continuous_scale='Purples')
fig2.show()
```

This map focuses on average salaries for female workers across countries from 
2024 to 2026. Higher salaries are concentrated in developed nations. Comparing 
this to the overall salary map highlights regions where gender pay gaps may be 
more pronounced, which is relevant given ongoing legal and social concerns 
about pay equity.

## Visualization 3: Automation Risk by Country Over Time

```python
risk_map = {'Low': 1, 'Medium': 2, 'High': 3, 'Very High': 4}
df['automation_risk_num'] = df['automation_risk_level'].map(risk_map)

scatter_data = df.groupby(['country', 'year'])['automation_risk_num'].mean().reset_index()
scatter_data['iso_alpha'] = scatter_data['country'].map(iso_map)
scatter_data = scatter_data.dropna(subset=['iso_alpha'])
scatter_data = scatter_data.sort_values('year')

fig3 = px.scatter_geo(scatter_data,
    locations='iso_alpha',
    color='automation_risk_num',
    hover_name='country',
    size='automation_risk_num',
    animation_frame='year',
    title='Average Automation Risk by Country Over Time (2024-2026)',
    color_continuous_scale='Reds',
    labels={'automation_risk_num': 'Automation Risk (1=Low, 4=Very High)'})
fig3.show()
```

This scatter geo map shows the average automation risk level of jobs by country 
from 2024 to 2026. Larger, darker circles indicate higher automation risk. 
This visualization helps identify which regions face the greatest workforce 
disruption from AI and automation over time. Countries with high automation 
risk should invest in workforce retraining programs to prepare workers for 
the future job market.
